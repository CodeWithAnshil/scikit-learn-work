```python
df.drop("body",axis=1,inplace=True)
print(f"Size of dataset after remove one column {df.shape}")
```

```python
from sklearn.impute import SimpleImputer
```

```python
print(f"Number of null values of column age before imputing: {df['age'].isnull().sum()}")
```

```python
impt=SimpleImputer(strategy='mean')
df['age']=impt.fit_transform(df[['age']])
print(f"Number of null values of column age after imputing: {df['age'].isnull().sum()}")
```

```python
def get_parameters(df):
    parameter={}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype=='float64' or df[col].dtype=='int64'or df[col].dtype=='int32':
            strategy='mean'
        else:
            strategy='most_frequent'

        missing_values=df[col][df[col].isnull()].values[0]
        parameter[col]={"missing_values":missing_values,"strategy":strategy}
    return parameter    

get_parameters(df)
```

```python
parameters=get_parameters(df)
for col, param in parameters.items():
    missing_values=param['missing_values']
    strategy=param['strategy']
    imp=SimpleImputer(missing_values=missing_values,strategy=strategy)
    df[col]=imp.fit_transform(df[[col]]).ravel()
    
```
