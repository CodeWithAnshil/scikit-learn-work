```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
mean = np.mean(df["total_bill"])
```

```python
std = np.std(df["total_bill"])
```

```python
mean, std
```

```python
# standardisation 
std_data = []
for i in df["total_bill"]:
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
sns.histplot(std_data, kde=True)
plt.xlabel("total_bill")
plt.title("After Appling standardisation")

plt.subplot(1, 2, 2)
sns.histplot(df["total_bill"], kde=True)
plt.title("Before Appling standardisation")
plt.show()
```
