```python
# Linear Regression models the relationship between dependent variable (y) and 
# one or more independent variable (x) by fitting a linear equation to observed data.
# In sample term if find best-fitting straight line through the data piont. 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataset load 
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.feature_names
```

```python
diabetes.target
```

```python
diabetes.data
```

```python
diabetes.DESCR
```

```python
# make dataframe 
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
df['target']=diabetes.target
```

```python
df
```

```python
# EDA, DataCleaning, Data Preparation, Feature Engineering
```

```python
# Divided data in X, y for simple linear regrassion we take only one features
X=df[['bmi']]
y =df['target']
```

```python
# split dataset using train test split with the help of sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train) #at backend, it will use gradiant descent to give optimal coefficients
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Visulisation of result 

plt.scatter(X_test, y_test, color='Blue', label='Actual Data')
plt.plot(X_test, y_pred, linewidth=3, color='green', label="Linear Regression Line")
plt.xlabel('BMI')
plt.ylabel('One year progression-target')
plt.title('Linear Regression on diabetes data')
plt.legend()
plt.show()
```
