```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification, make_regression
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, n_features=20, random_state=42)
```

```python
# split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Gradient Boosting Classifier
gradient  = GradientBoostingClassifier()
gradient
```

```python
# hyperparameter Truning 
from sklearn.model_selection import GridSearchCV
parameter = {
  'learning_rate': [0.1, 0.02, 0.04, 1, 0.4],
  'n_estimators': [20, 300, 400],
}
grid = GridSearchCV(gradient, param_grid=parameter, n_jobs=-1, verbose=3, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
model = grid.best_estimator_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Performace of Gradient Boosting Classification ")
print(f"Accuracy score : {accuracy_score(y_test, y_pred)}")
print(f"Confusion_matrix : {confusion_matrix(y_test, y_pred)}")
print(f"Classification_report : {classification_report(y_test, y_pred)}")
```

```python
# Gradient Boosting Regressor
X, y = make_regression(n_samples=1000, n_features=20, noise=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
grid_R =  GradientBoostingRegressor()
grid_R
```

```python
grid_r =model_r.best_estimator_
```

```python
y_pred = grid_r.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("Performence Prediction ")
print(f"Mean Absolute Error :  {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_test, y_pred)}")
print(f"r2_score : {r2_score(y_test, y_pred)}")
```
