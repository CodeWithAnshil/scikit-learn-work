```python
# Nominal/OneHotEncoder...

from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df1[['day', 'time']]).toarray()
```

```python
encoded_value = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_value
```

```python
encoded_value = encoded_value.drop(['day_Thur', 'time_Lunch'], axis=1)
```

```python
df1 = df1.drop(['day', 'time'], axis=1)
```

```python
new_df = pd.concat([df1, encoded_value], axis=1)
```
