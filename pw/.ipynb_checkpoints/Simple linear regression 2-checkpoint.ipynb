```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CSV/height_weight_data.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df[' Height (Inches)']
```

```python
# data divided into X and y 
X = df[[' Height (Inches)']]
y = df[[' Weight (Pounds)']]
```

```python
# split data using train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# scaling the data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
X_train
```

```python
# model learning
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
print(f"The slop of coef is {regression.coef_[0][0]}")
```

```python
y_pred = regression.predict(X_train)
y_pred
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred, linewidth=3, color='r')
plt.show()
```

```python
y_pred_test= regression.predict(X_test)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred_test, linewidth=3, color='black')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred_test) 
```

```python
mae = mean_absolute_error(y_test, y_pred_test) 
```

```python
rmse = np.sqrt(mse)
```

```python
mse, mae, rmse
```
