```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
beta = np.linspace(-5, 5, 200)
error_term = (beta-2)**2  # let as consider as actual coefficient= 2
penality_term = 5*(beta**2)
cost_F = error_term + penality_term
```

```python
plt.figure(figsize=(5,5))
plt.plot(beta, error_term, color='b', label='Error Term')
plt.plot(beta, penality_term, color='g', label='Penality Term')
plt.plot(beta, cost_F, color='r', label='Cost Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("This is graph between Beta, Error Term, Cost Function", fontsize=14, pad=20)
plt.legend()
plt.grid()
plt.show()
```
