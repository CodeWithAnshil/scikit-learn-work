```python
# data interpolation >>> is a process of estimating data b/w the given range
```

```python
# Linear interpolation
# Cubic interpolation
# Polynomial interpolation
```

```python
# Linear Interpolation

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 6, 8, 10, 12])
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x_new = np.linspace(1, 5, 10)
```

```python
y_interp = np.interp(x_new, x, y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
plt.show()
```

```python
# cubic interpolation

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from scipy.interpolate import interp1d 
```

```python
f = interp1d(x, y, kind='cubic',)
```

```python
x_new1 = np.linspace(1, 5, 50)
y_interp1 = f(x_new1)
```

```python
plt.scatter(x_new1, y_interp1)
plt.show()
```

```python
# poly interpolation 

x= np.array([1, 2, 3, 4, 5])
y= np.array([1, 4, 9, 1, 25])
```

```python
p = np.polyfit(x, y, 2)
```

```python
x_new = np.linspace(1, 5, 30)
y_interp = np.polyval(p, x_new)
```

```python
y_interp
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x_new, y_interp)
plt.show()
```
