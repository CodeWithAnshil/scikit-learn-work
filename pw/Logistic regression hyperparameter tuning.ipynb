```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.datasets import make_classification 

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=2)
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
param = {'penalty':('l1','l2', 'elasticnet'), 'C':[1, 2, 10, 30, 40]}
clf = GridSearchCV(model,param_grid=param, cv=5)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_clf = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred_clf))
print(classification_report(y_test, y_pred_clf))
print(accuracy_score(y_test, y_pred_clf))
```

```python
model = LogisticRegression(penalty='l2',C=1)
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Random Search Cross Vailidation
randomozer_clf = RandomizedSearchCV(model, param_distributions=param, cv=5, n_iter=10)
```

```python
randomozer_clf.fit(X_train, y_train)
```

```python
randomozer_clf.best_params_
```

```python
model = LogisticRegression(penalty='l2',C=10)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
