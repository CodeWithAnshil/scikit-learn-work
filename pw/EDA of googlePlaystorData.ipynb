```python
def remove_str(item):
    if str(item)[-1] == 'M':
        res=float(str(item).replace('M',''))
        res =res*1024
        return res
    elif str(item)[-1]== 'k':
        res=float(str(item).replace('k',''))
        return res
    elif str(item)[-1]== '+':
        res=float(str(item).replace('+',''))
        return res
    else:
        return str(np.NaN)
```

```python
def version_float(ver):
    try:
        if isinstance(ver, str):
            parts = ver.split('.')
            return float(parts[0]+'.'+''.join(parts[1:]))
        else:
            return ver
    except:
        return np.nan
```

```python
df_copy['Current Ver'] = df_copy['Current Ver'].apply(version_float)
```

```python
def clean_android_ver(v):
    if isinstance(v, str):
        # Step 1: replace 'Varies with device' with NaN
        if 'Veries' in v:
            return np.nan
        # Step 2: remove 'and up' or any text after it
        v = v.split('and')[0].strip()
        # Step 4: remove unwanted letters like 'W' (from '4.4W')
        v = v.split('W')[0].strip()
        # Step 3: if it's a range like '4.0 - 7.1.1', take the first part (min version)
        v = v.split('-')[0].strip()

        # convert dot into float friendly formate
        try:
            parts = v.split('.')
            return float(parts[0]+'.'+''.join(parts[1:]))
        except:
            return np.nan
    else:
        return v
```

```python
df_copy['Android Ver'] = df_copy['Android Ver'].apply(clean_android_ver)
```
