# SVM (Support Vector Machines Algorithm in Machine Learning)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
X=iris.data
y=iris.target
names=iris.target_names
names
```

```python
X.shape
```

```python
y.shape
```

```python
df=pd.DataFrame(X,columns=iris.feature_names)
df['species']=iris.target
df['species']=df["species"].replace(to_replace=[0,1,2],value=['setosa', 'versicolor', 'virginica'])
```

```python
df
```

```python
import seaborn as sns
sns.pairplot(data=df,hue='species',palette='cool')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',random_state=0)
svm.fit(X_train,y_train)
```

```python
pred=svm.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
rbf_svm=SVC(kernel='rbf',random_state=0)
rbf_svm.fit(X_train,y_train)
rbf_pred=rbf_svm.predict(X_test)
```

```python
rbf_pred
```

```python
accuracy_score(y_test,rbf_pred)
```
