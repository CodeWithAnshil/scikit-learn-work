```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv("level_salary_dataset.csv")
```

```python
dataset
```

```python
plt.figure(figsize= (5, 4))
sns.scatterplot(x="Level", y="Salary", data=dataset)
plt.show()
```

```python
x = dataset[["Level"]]
y = dataset["Salary"]
```

##### Change into the Polynomial Features with degree = 2

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100   # Accuracy of model 
```

```python
y_pred = lr.predict(x)  # Prediction on entire data
```

```python
plt.figure(figsize=(5, 4))
plt.plot(dataset["Level"], y_pred, c='r')
sns.scatterplot(x="Level", y="Salary", data=dataset)
plt.legend({"Predicted", "Original"})
plt.show()
```
