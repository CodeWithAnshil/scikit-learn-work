```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("placement_data.csv")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first', dtype = int)
```

```python
df[['placement_Yes']] = encoder.fit_transform(df[['Placement']])
```

```python
df.drop(df[['Placement']], axis=1)
```

```python
x = df[["Score", "CGPA"]]
# y = np.array()
y = df["placement_Yes"]
```

```python
x
```

```python
sns.scatterplot(x="CGPA", y="Score", data= df, hue= "placement_Yes")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
modelSvm = SVC(kernel='linear')
modelSvm.fit(x_train, y_train)
```

```python
modelSvm.score(x_test, y_test)*100
```

```python
modelSvm.score(x_train, y_train)*100
```
