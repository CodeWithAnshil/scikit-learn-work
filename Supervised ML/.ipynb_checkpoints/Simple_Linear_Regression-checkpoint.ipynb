```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Placement.csv")
dataset.head(4)
```

# Plot a graph for conformation that this is  Linear Regression or not

```python
plt.figure(figsize= (5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset, color='r')
plt.show()
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred= lr.predict(x)
```

```python
plt.figure(figsize= (5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset, color='r')
plt.plot(dataset["cgpa"],y_pred,c='b')
plt.legend(["Original Data","Predict Data"])
plt.show()
```

```python
lr.score(x_test,y_test)*100  # accurancy of prediction
```

```python
lr.coef_[0]  # coeffiecient of linear equation
```

```python
lr.intercept_  # intercept point of linear quation
```
