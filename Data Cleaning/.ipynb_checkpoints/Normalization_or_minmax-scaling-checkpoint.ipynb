```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("train.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
mms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"] = mms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(4)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before Normalization")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1,2,2)
plt.title("After Normalization")
sns.histplot(dataset["CoapplicantIncome_min"], kde=True)
plt.show()
```
