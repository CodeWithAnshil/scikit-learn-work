```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Train.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(10, 5))
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

## Using the concept of Function Transformer

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_ft"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize= (10, 5))
plt.subplot(1, 2, 1)
plt.title("Befor")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_ft"])
plt.show()
```

### Remove Outliers 

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
min_range, max_range
```

```python
dataset = dataset[dataset["CoapplicantIncome"] <= max_range]
```

```python
plt.figure(figsize=(10, 5))
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```
