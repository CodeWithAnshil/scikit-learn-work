# Replace missing value with mode of that columns

```python
import pandas as pd 
```

```python
dataset=pd.read_csv("Diabetes Missing Data.csv")
dataset.head(3)
```

```python
dataset.dtypes
```

```python
dataset["Skin_Fold"].fillna(dataset["Skin_Fold"].mode()[0],inplace=True)
```

```python
for i in dataset.select_dtypes(include="float64").columns:
    dataset[i].fillna(dataset[i].mode()[0])
```
